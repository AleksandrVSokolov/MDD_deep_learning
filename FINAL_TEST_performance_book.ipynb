### Generate AUC figures for final tests

```python
import os
import pandas
import re
from HelperScriptsPY import helper_functions

base_dir = "RESULTS_FINAL_TEST"
ml_files = [base_dir + "/" + x for x in os.listdir(base_dir) if not re.search(".xlsx", x)]

merged_files = []

for x in ml_files:
    dataset = pandas.read_csv(x)
    dataset = helper_functions.drop_useless_column(dataset)
    dataset.replace({
        'model': {
            "SVM_c_RBF_15": "SVM (RBF kernel)",
            "SVM_c": "SVM (Linear kernel)",
            "RidgeLogisticRegr": "Logistic regression ('L2')",
            "RandomForest": "Random forest",
            "LogisticRegr": "Logistic regression (no regularization)",
            "Lasso_classifier": "Logistic regression ('L1')",
            "Elastic_net_classifier": "Logistic regression ('L1L2'/Elastic net)",
            "DecisionTree": "Decision tree"},
            'cpg_type': {
            "ALL":"Top 200 Limma",
            "Top_200_all_consistent_CpGs.txt": "Top 200 Consistent (Pre-selected)"}
    }, inplace=True)
    dataset = dataset[["model", 
                       "mval",
                       "accuracy_train",
                       "accuracy_test",
                       "auc_train",
                       "auc_test",
                       "cpg_type",
                       "path"]]
    merged_files.append(dataset)
merged_files = pandas.concat(merged_files)
merged_files.loc[:, merged_files.dtypes != "object"] = merged_files.loc[:, merged_files.dtypes != "object"].round(decimals=3)

# Selecting only betas for models (except for VAE)
merged_files = merged_files[merged_files["mval"] == "Mval"]
merged_files.reset_index(inplace=True, drop=True)
for i in range(merged_files.shape[0]):
    if merged_files["cpg_type"].values[i] == "Top 10K Limma":
        merged_files.loc[i, ("model")] = merged_files.loc[i, ("model")] + " 10K"
        
merged_files_limma = merged_files[merged_files["cpg_type"].isin(["Top 200 Limma", "Top 10K Limma"])]
merged_files_bias = merged_files[~merged_files["cpg_type"].isin(["Top 200 Limma", "Top 10K Limma"])]
merged_files_bias
```

```python
# Code to make ROC curve for biased feature models
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve (positive bias, harmonized data)')
for i in range(merged_files_bias.shape[0]):
    path = merged_files_bias.path.values[i]
    path = path.split("/")
    path = path[0:(len(path)-1):1]
    path = '/'.join(path)
    tpr_file = [path + "/" + file for file in os.listdir(path) if re.search("_tpr_test", file)]
    fpr_file = [path + "/" + file for file in os.listdir(path) if re.search("_fpr_test", file)]

    fpr = pandas.read_csv(fpr_file[0], header=None)
    tpr = pandas.read_csv(tpr_file[0], header=None)

    fpr = helper_functions.drop_useless_column(fpr)
    tpr = helper_functions.drop_useless_column(tpr)

    auc = merged_files_bias.auc_test.values[i]
    model = merged_files_bias.model.values[i]

    plt.plot(fpr[1], tpr[1], label= f"{model} ({auc})")
plt.legend(loc='best')
plt.savefig("ROC_curve_final_test_bias.pdf", dpi=300)
```
