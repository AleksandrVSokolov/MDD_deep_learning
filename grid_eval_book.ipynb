### Performance of small DNN classifiers

```python
import os
import pandas
from HelperScriptsPY.helper_functions import drop_useless_column

os.makedirs("Grid_search_results", exist_ok=True)

base_dir = "test_grid_simpleDNN_BIANCA"

subfolders = [base_dir + "/" + x for x in os.listdir(base_dir) if os.path.isdir(base_dir + "/" + x)]

performance_lists = []
for subfolder in subfolders:
    path = subfolder + "/" + "average_stats.csv"
    dataset = pandas.read_csv(path)
    dataset = drop_useless_column(dataset)
    dataset["path"] = path
    performance_lists.append(dataset)

performance_df = pandas.concat(performance_lists)
performance_df = performance_df.sort_values(by=["auc_test"], ascending=False)
performance_df.to_csv("Grid_search_results/SmallDNNclassif.csv")
performance_df
```

The analysis produces comparable with ML accuracy, eaching AUCs of 0.56 (with unbiased features)
